**Analytic Approach**

Langkah pertama dalam membangun model prediksi harga rumah yang akurat adalah melakukan eksplorasi dan pra-pemrosesan data. Proses ini mencakup pembersihan data, pemahaman terhadap fitur-fitur yang ada, serta analisis hubungan antara fitur-fitur tersebut dengan harga rumah. Setelah itu, model regresi akan dibangun dan dilatih untuk memprediksi harga rumah berdasarkan pola yang ditemukan dalam data. Berikut adalah tahapan yang akan dilakukan:

1. Eksplorasi Data: Eksplorasi data dilakukan untuk memahami distribusi data serta hubungan antara fitur-fitur dalam dataset (seperti jumlah kamar, lokasi, dan pendapatan median) dengan target (harga rumah). Tujuannya adalah untuk memahami pola-pola yang ada dan mengidentifikasi fitur-fitur mana yang memiliki pengaruh paling besar terhadap harga rumah.

2. Preprocessing Data:
    - Handling Missing Values: Nilai yang hilang akan diisi menggunakan nilai median atau teknik imputation lainnya.
    - Encoding Kategorikal: Fitur kategorikal (misalnya, ocean_proximity) akan dikonversi menjadi format numerik menggunakan One-Hot Encoding untuk memastikan model dapat memprosesnya dengan benar.
    - Skalasi Fitur: Fitur dengan skala berbeda akan dinormalisasi agar model dapat berfungsi dengan baik.

3. Pemilihan Model Regresi: Setelah data dipersiapkan, beberapa model regresi akan dipilih untuk diuji dan dibandingkan. Model regresi linear seperti Linear Regression akan menjadi model pertama yang diuji. Selain itu, model regresi non-linear seperti Decision Tree Regression dan Random Forest Regression akan dipertimbangkan untuk mengevaluasi apakah model yang lebih kompleks dapat memberikan hasil yang lebih akurat. Pemilihan model dilakukan berdasarkan kemampuannya dalam memprediksi harga rumah sesuai dengan pola-pola yang ditemukan dalam data.

4. Pelatihan Model dan Evaluasi: Setelah model dipilih, model akan dilatih menggunakan data latih dan diuji untuk melakukan prediksi harga rumah pada data uji. Kinerja model akan dievaluasi menggunakan metrik seperti RMSE, MAE, dan MAPE untuk mengukur akurasi prediksi model.

**Metric Evaluation**

Untuk mengevaluasi kinerja model regresi, beberapa metrik akan digunakan untuk memberikan gambaran tentang seberapa baik model dalam memprediksi harga rumah. Berikut adalah metrik yang akan digunakan:

1. RMSE (Root Mean Squared Error): Mengukur rata-rata kesalahan kuadrat antara prediksi dan nilai aktual. Semakin kecil nilai RMSE, semakin baik model dalam memprediksi harga rumah.
2. MAE (Mean Absolute Error): Mengukur rata-rata kesalahan absolut antara prediksi dan nilai aktual. Metrik ini memberikan gambaran langsung tentang kesalahan rata-rata yang dibuat oleh model.
3. MAPE (Mean Absolute Percentage Error): Mengukur rata-rata persentase kesalahan prediksi. Nilai MAPE yang lebih kecil menunjukkan prediksi harga rumah yang lebih akurat dalam konteks persentase.

Semakin kecil nilai RMSE, MAE, dan MAPE yang dihasilkan, berarti model semakin akurat dalam memprediksi harga sewa sesuai dengan limitasi fitur yang digunakan. 

Selain itu jika model yang terpilih adalah linear model, maka  R-squared atau adj. R-squared dapat digunakan sebagai metric evaluation.

4. R-squared (R²): Mengukur seberapa baik model dapat menjelaskan variasi data. Nilai yang mendekati 1 menunjukkan model yang baik, karena semakin besar nilai R², semakin tepat model memprediksi harga.
5. Adjusted R-squared: Mengoreksi nilai R² dengan mempertimbangkan jumlah fitur dalam model, sehingga lebih berguna ketika membandingkan model dengan jumlah fitur yang berbeda.

```python
# Define Algo
linear = LinearRegression()
tree = DecisionTreeRegressor(random_state=0)
knn = KNeighborsRegressor()
rf = RandomForestRegressor(random_state=0)
xgb = XGBRegressor(random_state=0)

list_model = [linear, tree, knn, rf, xgb]

scorer = ["neg_root_mean_squared_error",
          "neg_mean_absolute_percentage_error",
          "neg_mean_absolute_error"]

# List Score RMSE
list_score_mean_rmse = []
list_score_std_rmse = []

# List Score MAE
list_score_mean_mae = []
list_score_std_mae = []

# List Score MAPE
list_score_mean_mape = []
list_score_std_mape = []

# Cross Validation loop
for model in list_model:
    pipe_model = Pipeline([
        ("Preprocessing", transformer),
        ("Modeling", model)])

    cv_score = cross_validate(
        estimator=pipe_model,
        X=X_train,
        y=y_train,
        cv=5,
        scoring=scorer,
        return_train_score=True)

    # Scoring RMSE
    list_score_mean_rmse.append(cv_score["test_neg_root_mean_squared_error"].mean())
    list_score_std_rmse.append(cv_score["test_neg_root_mean_squared_error"].std())

    # Scoring MAE
    list_score_mean_mae.append(cv_score["test_neg_mean_absolute_error"].mean())
    list_score_std_mae.append(cv_score["test_neg_mean_absolute_error"].std())

    # Scoring MAPE
    list_score_mean_mape.append(cv_score["test_neg_mean_absolute_percentage_error"].mean())
    list_score_std_mape.append(cv_score["test_neg_mean_absolute_percentage_error"].std())
    
```

```python
df_cv = pd.DataFrame()
df_cv["Model"] = ['Linear Regression', 'DecisionTree Regressor', 'KNN Regressor', 'RandomForest Regressor', 'XGBoost Regressor']
df_cv["Mean RMSE"] = list_score_mean_rmse
df_cv["Std RMSE"] = list_score_std_rmse
df_cv["Mean MAE"] = list_score_mean_mae
df_cv["Std MAE"] = list_score_std_mae
df_cv["Mean MAPE"] = list_score_mean_mape
df_cv["Std MAPE"] = list_score_std_mape
df_cv
```

Berdasarkan hasil cross-validation, XGBoost menunjukkan performa yang lebih unggul dengan nilai mean RMSE, MAE, dan MAPE yang lebih kecil dibandingkan RandomForest. Ini menunjukkan bahwa prediksi dari XGBoost lebih akurat. Namun jika dilihat lebih cermat, RandomForest memiliki nilai standard deviation (Std) yang lebih kecil pada RMSE dan MAPE, yang menunjukkan bahwa model ini sedikit lebih stabil dan konsisten dalam hasil prediksinya. 

Uuntuk memastikan model terbaik, maka diputuskan melakukan uji coba (benchmarking) pada test set untuk mengevaluasi kinerja kedua model dengan data yang belum pernah dilihat sebelumnya dan memilih model yang lebih baik dalam memprediksi data baru dengan akurat.

**Predict to Test Set with Random Forest & XGBoost**

```python
# Benchmark 2 model terbaik
models = {
    'RandomForest': RandomForestRegressor(random_state=1),
    'XGB': XGBRegressor(random_state=1)  
}

score_rmse = []
score_mae = []
score_mape = []

# Loop untuk menguji kedua model
for i in models:
    model = Pipeline([
        ('preprocessing', transformer),
        ('model', models[i])
        ])

    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    score_rmse.append(np.sqrt(mean_squared_error(y_test, y_pred)))
    score_mae.append(mean_absolute_error(y_test, y_pred))
    score_mape.append(mean_absolute_percentage_error(y_test, y_pred))

score_before_tuning = pd.DataFrame({'RMSE': score_rmse, 'MAE': score_mae, 'MAPE': score_mape}, index=models.keys())
score_before_tuning
```

Berdasarkan hasil benchmarking antara model XGBoost dan Random Forest, dapat disimpulkan bahwa XGBoost menghasilkan nilai error (RMSE, MAE, dan MAPE) yang lebih kecil dibandingkan dengan Random Forest, yang menunjukkan akurasi prediksi yang lebih tinggi. Dengan demikian, XGBoost dipilih sebagai model terbaik untuk melakukan modeling pada project ini.

### **5. Hyperparameter Tuning**

```python
# Define Hyperparameter
hyperparameter = {
    'Modeling__n_estimators': list(np.arange(100, 1001, 100)), # Jumlah pohon 100 hingga 1000 dengan kelipatan 100
    'Modeling__max_depth': list(np.arange(3, 22)),           # Kedalaman pohon 3 hingga 21
    'Modeling__learning_rate': list(np.arange(1, 100)/100),  
    'Modeling__subsample': list(np.arange(2, 11)/10),        # Jumlah baris tiap pohon (% dari total baris train set)
    'Modeling__colsample_bytree': list(np.arange(1, 11)/10), # Jumlah feature yang digunakan untuk tiap pohon (% dari total kolom train set)
    'Modeling__reg_alpha': list(np.logspace(-3, 1, 10))      # 0.001 hingga 10 untuk regularisasi L1. # Semakin besar nilainya, semakin konservatif/simpel modelnya
   }

# Define Model
xgb_model = XGBRegressor(random_state=0)

# Define Scorer
scorer = ["neg_root_mean_squared_error",
          "neg_mean_absolute_percentage_error",
          "neg_mean_absolute_error"
]

# Define pipeline with preprocessing and the XGBoost model
pipe_model = Pipeline([
    ("Preprocessing", transformer),
    ("Modeling", xgb_model)
])

# Setup RandomizedSearchCV with cross-validation
randomized_xgb = RandomizedSearchCV(
    estimator=pipe_model,
    param_distributions=hyperparameter,
    n_iter=100, 
    scoring=scorer, 
    refit="neg_root_mean_squared_error",  
    cv=5,  
    n_jobs=-1,  
    verbose=1,
    random_state=0
)
```

```python
# Fit the model to the training data
randomized_xgb.fit(X_train, y_train)
```
